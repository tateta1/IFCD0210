## 1. Arquitecturas y Protocolos
Este bloque sienta las bases teóricas de cómo se comunican las aplicaciones en red.

- Modelos de arquitectura: Análisis de arquitecturas de n-capas (cliente, servidor, datos) y el papel del middleware.
- Protocolos de comunicación: Profundización en HTTP/HTTPS y otros protocolos de nivel de aplicación necesarios para el intercambio de datos.
- Seguridad en la comunicación: Conceptos básicos de cifrado, certificados SSL y túneles de comunicación segura.

## 2. Programación en el Servidor
Es el núcleo técnico de la unidad, donde se desarrolla la lógica de la aplicación.

- Lenguajes y entornos: Selección y configuración del entorno de ejecución (como Node.js, PHP, .NET o Java).
- Gestión de peticiones y respuestas: Cómo recibir datos del cliente, procesarlos y devolver resultados (cabeceras, cuerpos de respuesta y códigos de estado).
- Sesiones y estado: Técnicas para mantener la información del usuario entre diferentes peticiones (cookies, tokens JWT y variables de sesión).

## 3. Servicios Web y APIs (Interoperabilidad)
Se enfoca en que la aplicación pueda "hablar" con otros sistemas externos.

- Arquitectura REST: Diseño de APIs basadas en recursos, utilizando métodos estándar (GET, POST, PUT, DELETE).
- Formatos de intercambio: Uso de JSON y XML para el envío de datos estructurados entre aplicaciones.
- Consumo de servicios: Cómo realizar peticiones desde nuestro servidor a APIs de terceros (pasarelas de pago, mapas, redes sociales).

## 4. Gestión de la Persistencia en Entornos Distribuidos
Trata sobre cómo guardar y recuperar información de forma eficiente.

- Acceso a bases de datos: Implementación de conexiones y consultas desde el código del servidor.
- Modelos de datos distribuidos: Introducción a cómo se gestionan los datos cuando no están en un único servidor (replicación y alta disponibilidad).
- Optimización: Uso de sistemas de caché para mejorar la velocidad de respuesta de las aplicaciones distribuidas.
