## 1. Introducción a la arquitectura n-capas

La **arquitectura n-capas (n-tier)** es un modelo de diseño de software utilizado en sistemas distribuidos donde la aplicación se divide en múltiples niveles o capas independientes. Cada capa tiene una responsabilidad específica y se comunica con las capas adyacentes. 

Este enfoque es común en aplicaciones web y sistemas empresariales, donde se separan la interfaz de usuario, la lógica de negocio y la gestión de datos. Esta separación permite modificar una capa sin afectar necesariamente al resto del sistema. :contentReference[oaicite:3]{index=3}

El modelo más habitual es la **arquitectura de tres capas (three-tier)**:

```
Cliente (Presentación)
        │
Servidor de Aplicaciones (Lógica de negocio)
        │
Servidor de Datos (Base de datos)
```

La arquitectura n-capas mejora la **escalabilidad, mantenibilidad y seguridad**, ya que cada capa puede desarrollarse y desplegarse de forma independiente. :contentReference[oaicite:4]{index=4}

## 2. Capas principales de la arquitectura

### 2.1 Capa de presentación (cliente)

Es la capa con la que interactúa el usuario final. Su función principal es mostrar información y recoger datos de entrada.

Ejemplos:
- Navegador web
- Aplicación móvil
- Interfaz gráfica

Responsabilidades:
- Mostrar datos al usuario
- Enviar solicitudes al servidor
- Validación básica de datos


````python
# Simulación de una capa cliente

import requests

def obtener_usuario(id_usuario):
    url = f"http://localhost:5000/usuarios/{id_usuario}"
    respuesta = requests.get(url)
    return respuesta.json()

usuario = obtener_usuario(1)
print(usuario)
````

### 2.2 Capa de lógica de negocio (servidor de aplicaciones)

Esta capa contiene las **reglas de negocio** y la lógica que procesa las solicitudes del cliente.

Funciones principales:
- Procesar solicitudes
- Aplicar reglas del sistema
- Comunicarse con la capa de datos

Ejemplos de tecnologías:
- APIs REST
- Microservicios
- Servidores de aplicaciones

En aplicaciones web modernas esta capa suele implementarse mediante APIs.

````python
# Ejemplo de servidor de aplicaciones usando Flask

from flask import Flask, jsonify

app = Flask(__name__)

usuarios = {
    1: {"nombre": "Ana", "edad": 25},
    2: {"nombre": "Carlos", "edad": 30}
}

@app.route('/usuarios/<int:id_usuario>')
def obtener_usuario(id_usuario):
    usuario = usuarios.get(id_usuario)
    return jsonify(usuario)

if __name__ == '__main__':
    app.run(debug=True)
````

### 2.3 Capa de datos

La capa de datos es responsable del **almacenamiento y recuperación de información**.

Incluye:
- Bases de datos
- Sistemas de almacenamiento
- Sistemas de caché

Esta capa suele estar protegida y no es accesible directamente desde el cliente.

````python
# Ejemplo simple de acceso a datos con SQLite

import sqlite3

def obtener_usuario_db(id_usuario):
    conexion = sqlite3.connect("usuarios.db")
    cursor = conexion.cursor()

    cursor.execute("SELECT nombre, edad FROM usuarios WHERE id=?", (id_usuario,))
    resultado = cursor.fetchone()

    conexion.close()

    if resultado:
        return {"nombre": resultado[0], "edad": resultado[1]}
    return None
````

## 3. El papel del middleware

El **middleware** es un software intermedio que facilita la comunicación entre diferentes componentes o capas del sistema.

Actúa como una especie de "puente" entre aplicaciones, bases de datos y servicios.

Funciones del middleware:
- Gestión de comunicación entre servicios
- Autenticación y seguridad
- Balanceo de carga
- Gestión de transacciones
- Integración entre sistemas

En arquitecturas n-tier el middleware suele encontrarse en la **capa intermedia (application tier)** y puede gestionar conexiones entre clientes y bases de datos, permitiendo que múltiples clientes compartan recursos del servidor. :contentReference[oaicite:5]{index=5}

````python
# Ejemplo conceptual de middleware en Python

def middleware_autenticacion(funcion):
    def wrapper(usuario):
        if usuario != "admin":
            return "Acceso denegado"
        return funcion(usuario)
    return wrapper

@middleware_autenticacion
def obtener_datos(usuario):
    return "Datos sensibles del sistema"

print(obtener_datos("admin"))
````

## 4. Ventajas de la arquitectura n-capas

- **Escalabilidad:** cada capa puede escalarse de forma independiente.
- **Mantenibilidad:** los cambios en una capa no afectan necesariamente a las demás.
- **Seguridad:** la capa de datos puede aislarse del acceso directo del cliente.
- **Reutilización:** la lógica de negocio puede ser utilizada por diferentes interfaces.

La separación de responsabilidades facilita el mantenimiento y evolución del sistema. :contentReference[oaicite:6]{index=6}

## 5. Ejemplo completo de flujo en arquitectura de tres capas

```
Usuario → Cliente (Frontend)
        → API / Servidor (Backend)
        → Base de datos
```

Flujo:
1. El cliente solicita información.
2. El servidor procesa la solicitud.
3. El servidor consulta la base de datos.
4. La respuesta vuelve al cliente.
